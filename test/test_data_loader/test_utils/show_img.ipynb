### Visualize segmentation result

```python
import matplotlib.pyplot as plt
import numpy as np
import os

from PIL import Image
```

```python
def visualize_img(image_path: str, index: int):
    # VISUALIZE_MAPPER = {
    #     0: 2, # 
    #     1: 8, # 
    #     2: 16, # 
    #     3: 32, # 
    # }
    VISUALIZE_MAPPER = np.array([[0, 64, 64], [64, 0, 64], [64, 64, 0], [64, 64, 64]], dtype=np.uint8)

    original_path = os.path.join(image_path, f'{index}.png')
    original_img = Image.open(original_path)

    mask_path = os.path.join(image_path, f'{index}_class.png')
    mask_img = Image.open(mask_path)

    original_img = np.array(original_img)
    mask_img = np.array(mask_img)[:, :, 0]
    mapped_mask_img = VISUALIZE_MAPPER[mask_img]
    plt.imshow(original_img)
    plt.imshow(mapped_mask_img, alpha=0.7)
    plt.show()
```

```python
visualize_img('/opt/dataset/tr2_cropped/', 10)
```
